# Problem 6

Determine how many males and females won the prize. Does the sum of male and female prizes add up to the total entries? If not, why?

On the same graph make histograms for the numbber of males and females that won the prize each year. 

```python

print(df['sex'].value_counts(dropna=False))

total_prizes = len(df)
known_sex_prizes = df['sex'].count()  
print(f"Total Prizes: {total_prizes}, Sex entries: {known_sex_prizes}")


df['decade'] = (df['year'] // 10) * 10
grouped = df.groupby(['decade', 'sex']).size().unstack(fill_value=0)
all_decades = np.arange(df['decade'].min(), df['decade'].max() + 10, 10)


fig, ax = plt.subplots(figsize=(10, 6))


x = grouped.index 
width = 5  

ax.bar(x + width , grouped['Male'], width=width, color='cornflowerblue', label='Male')
ax.bar(x + width / 2, grouped['Female'], width=width, color='pink', label='Female')

ax.set_xticks(all_decades)
ax.set_xticklabels(all_decades, rotation=45)
ax.set_xlabel('  Year')
ax.set_ylabel('Number of Winners')
ax.set_title('Male and Female Nobel Prize Winners per Decade')
ax.legend()

plt.show()
```

# Problem 8

Determine the age that each of the prize winners was when they won the prize. Find the youngest and oldest to win the prize. Find the mean age to win the prize, and the standard deviation of it. Make a histogram of the age each person was when they won the prize.

### Hint: You can use Pandas DatetimeIndex method to extract the year from the date format yyyy-mm-dd
> pd.DatetimeIndex(date).year

Where *date* is given in a formate like year-month-day, the above will returen the year, hence the *.year* in the call above. Use 'shit+tab' on pandas.DatetimeIndex to learn more.

```python
df['age_at_award'] = df['year'] - pd.DatetimeIndex(df['birth_date']).year
df = df[df['age_at_award'].notna()]
#print(df.isna().sum())
youngest = df.loc[df['age_at_award'].idxmin()]
oldest = df.loc[df['age_at_award'].idxmax()]
#print(youngest)

print(f"Youngest Winner: {youngest['full_name']} at age {youngest['age_at_award']} in {youngest['year']}")
print(f"Oldest Winner: {oldest['full_name']} at age {oldest['age_at_award']} in {oldest['year']}")

mean_age = df['age_at_award'].mean()
std_age = df['age_at_award'].std()
print(f"Mean Age: {mean_age:.2f}")
print(f"Standard Deviation of Age: {std_age:.2f}")



fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(df['age_at_award'], bins=20, color='skyblue', edgecolor='black')
ax.set_xlabel('Age at Award')
ax.set_ylabel('Number of Winners')
ax.set_title('Distribution of Age of Nobel Prize Winners')
print()
```
